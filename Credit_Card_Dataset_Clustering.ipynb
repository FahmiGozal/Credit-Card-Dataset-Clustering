```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.preprocessing import StandardScaler, normalize
from sklearn.model_selection import train_test_split
import seaborn as sns
```

```python
df = pd.read_csv('CC GENERAL.csv')
```

```python
df.isnull().any()
```

```python
from sklearn.cluster import KMeans
x_train, x_test, y_train, y_test = train_test_split(df['BALANCE'].values, df['CREDIT_LIMIT'].values, test_size = 0.2, random_state=0)
#points = zip(x_train,y_train)
points = np.column_stack((x_train, y_train))
plt.scatter(x_train,y_train)

model = KMeans(n_clusters = 4)
model.fit(points)
labels = model.predict(points)
plt.scatter(x_train,y_train,c=labels)
```
